```python
def query_preprocessing(query_results_list):
    cars_dataset = {}

    for i in query_results_list:
        #cars_dataset["id"]=
        general_data={"Condition":str(i["condition"]),
                        "State":str(i["location"]["state"]["name"]),
                        "City":str(i["location"]["city"]["name"]),
                        "Price":float(i["price"]),
                        "Currency":str(i["currency_id"])}

        car_attributes={}
        for attr in i["attributes"]:
            car_attributes={**car_attributes,**{attr["id"]:attr["value_name"]}}    

        cars_dataset[i["id"]]={**general_data,**car_attributes}
    
    return cars_dataset
    #return pd.DataFrame(cars_dataset).T
```
