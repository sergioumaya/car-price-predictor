```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [20, 12]

stats = open("/home/sergio/Documents/stats.txt","r")

def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

pom_5v_in = []
pom_5v_gpu = []
pom_5v_cpu = []

for i in stats:
    #print(i)
    start_1 = i.find("POM_5V_IN")
    start_2 = i.find("POM_5V_GPU")
    start_3 = i.find("POM_5V_CPU")
    
    pom_5v_in.append( float(i[start_1+10:start_1+20].split("/")[0]))
    pom_5v_gpu.append(float(i[start_2+10:start_2+20].split("/")[0]))
    pom_5v_cpu.append(float(i[start_3+10:start_3+20].split("/")[0]))
    
mov_pom_5v_in = moving_average(pom_5v_in,20)
mov_pom_5v_gpu = moving_average(pom_5v_gpu,20)
mov_pom_5v_cpu = moving_average(pom_5v_cpu,20)
mov_total = mov_pom_5v_in+mov_pom_5v_gpu+mov_pom_5v_cpu

total = np.array(pom_5v_in)+np.array(pom_5v_gpu)+np.array(pom_5v_gpu)

'''
plt.figure()    
plt.plot(pom_5v_in,label="POM 5V IN")
plt.plot(pom_5v_gpu,label="POM 5V GPU")
plt.plot(pom_5v_cpu,label="POM 5V CPU")
#plt.plot(total,label="TOTAL")
plt.ylim([0,7500])
plt.grid()
plt.xlabel("Time(s)")
plt.ylabel("Power (mW)")
plt.legend()
'''


plt.figure()    
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#plt.xlabel(fontsize=14)
plt.plot(mov_pom_5v_in/1000,label="TOTAL POM")
plt.plot(mov_pom_5v_gpu/1000,label="GPU POM")
plt.plot(mov_pom_5v_cpu/1000,label="CPU POM")
#plt.plot(mov_total,label="TOTAL")
plt.ylim([0,7])
plt.xlim([0,1120])
plt.grid()
plt.xlabel("Time (s)",fontsize=14)
plt.ylabel("Power Consumption (W)",fontsize=14)
plt.legend()
```
